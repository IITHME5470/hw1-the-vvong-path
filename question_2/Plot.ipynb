```python
import matplotlib.pyplot as plt
import re

with open('timetaken.log', 'r') as file:
    lines = file.readlines()

n_values = []
time_taken_values = []
pattern = r'n = (\d+), avg time taken = (\d+\.\d+)'

# Extract data using the pattern
for line in lines:
    match = re.search(pattern, line)
    if match:
        n = int(match.group(1))
        time_taken = float(match.group(2))
        n_values.append(n)
        time_taken_values.append(time_taken)

# Plotting the data
plt.plot(n_values, time_taken_values, marker='o', linestyle='-', color='b')
plt.xlabel('n')
plt.ylabel('Avg Time Taken (seconds)')
plt.title('n vs Time Taken')
plt.grid(True)
plt.show()
```
